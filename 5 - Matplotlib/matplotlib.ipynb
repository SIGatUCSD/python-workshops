### 5.1 What is Matplotlib and why do we use it?

Matplotlib is a Python library designed for creating visualizations. It can be used in a variety of interfaces, such as IPython shells, notebooks and makes conducting data analysis easier overall. While it is a great library, Matplotlib can get *very*, *very* complicated and as such we will only be covering the basics.

```python
import matplotlib.pyplot as plt # this is the standard convention of importing matplotlib
import numpy as np
import pandas as pd
```

### 5.2 Array-like: list, NumPy arrays, pandas Series and creating our first plot

When we pass data in for our arguments in Matplotlib, they usually need to be array-like. Think of it as anything that could be interpreted as an array such as a numpy array, a scalar (number), Python list, pandas Series, etc.

This link shows a good visual of the components of a plot: https://matplotlib.org/stable/tutorials/introductory/quick_start.html

```python
# examples of array-likes
arr = np.array([1, 2, 3, 4, 5])
ls = [2, 4, 6, 8, 10]
```

```python
# we can create objects in Matplotlib
fig, ax = plt.subplots() # the figure is what our plots will be shown on and an ax is an axis
ax.plot(arr, ls) # example of a simple line chart
```

### 5.3 Customizing our plots

```python
plt.show() # this is standard convention of showing our plots, but in a notebook they show by default
```

Matplotlib offers several options to customize your plot. In this section, we will briefly cover some of them.

```python
fig # what our plot currently looks like for reference
```

```python
ax.set_title("Line Chart") # sets a title for the axis
ax.set_ylabel("Y") # sets a y-axis label for the axis
ax.set_xlabel("X") # sets an x-axis label for the axis
ax.set_ylim(0, 12) # sets bounds for the limits
fig # what our new plot looks like
```

```python
fig, ax = plt.subplots()
ax.scatter(arr, ls, color="black") # specify the color of the points
ax.set(                            # a way to set multiple arguments at once
    title="Scatter Plot",
    xlabel="X",
    ylabel="Y"
)
```

### 5.4 Other plots

```python
df = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data", header=None) # a test dataset from the uci machine learning repository
```

```python
df.head() # what our dataframe looks like
```

```python
fig, ax = plt.subplots()
ax.hist(df[1], color="skyblue") # creating a histogram in matplotlib
```

```python
fig, ax = plt.subplots() # creating a pie chart in matplotlib
ax.pie(df[0].value_counts(normalize=True), labels=df[0].unique())
ax.set_title("Pie Chart")
ax.set_xlabel("Categories")
```

One of the easiest ways to make our graph look nicer is using plt.style(). As you will see in the example below, there will be changes to the background and axes, but you could also configure these manually.

```python
plt.style.use("ggplot")
data = {"O": 60, "AAPL": 160, "SPY": 400}
fig, ax = plt.subplots()
ax.bar(data.keys(), data.values(), color="pink") # example of a barchart
ax.set_title("Bar Chart")
ax.set_ylabel("Stock Price ($)")
```
